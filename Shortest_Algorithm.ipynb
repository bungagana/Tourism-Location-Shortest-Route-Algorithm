```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import time

# Inisialisasi
INF = float('inf')

# Input jumlah node
num_nodes = int(input("Masukkan banyak node (jumlah lokasi wisata): "))

# Input nama node
nodes = []
print("\nMasukkan nama-nama node (contoh: Stasiun, Alun-Alun, Museum):")
for i in range(num_nodes):
    node = input(f"Nama Node {i+1}: ")
    nodes.append(node)

# Inisialisasi graf
edges = []
print("\nMasukkan jarak antar node (contoh: 3 untuk 3 km). Jika tidak ada jalur langsung, lewati.")
for i in range(num_nodes):
    for j in range(i+1, num_nodes):
        jarak = input(f"Jarak dari {nodes[i]} ke {nodes[j]} (jika tidak ada, kosongkan): ")
        if jarak:
            jarak = int(jarak)
            edges.append((i, j, jarak))
            edges.append((j, i, jarak))  # Graf tidak berarah

# Algoritma Bellman-Ford dengan Path Tracking
def bellman_ford(edges, num_nodes, start):
    # Inisialisasi jarak dan jalur
    dist = [INF] * num_nodes
    dist[start] = 0
    path = [-1] * num_nodes  # Menyimpan node sebelumnya

    # Relaksasi edge (V-1) kali
    for _ in range(num_nodes - 1):
        for u, v, weight in edges:
            if dist[u] != INF and dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight
                path[v] = u  # Simpan jalur

    # Deteksi negative cycle (opsional)
    for u, v, weight in edges:
        if dist[u] != INF and dist[u] + weight < dist[v]:
            print("Graf mengandung negative cycle!")
            return None, None

    return dist, path

# Fungsi untuk Melacak Jalur Terpendek
def get_path(path, start, end):
    route = []
    if start == end:
        route.append(start)
    elif path[end] == -1:
        return route  # Tidak ada jalur
    else:
        while end != -1:
            route.append(end)
            end = path[end]
        route.reverse()
    return route

# Pilih titik awal untuk Bellman-Ford
print("\nDaftar Node:")
for idx, name in enumerate(nodes):
    print(f"{idx}. {name}")

start_node = int(input("\nMasukkan nomor titik awal: "))

# Menghitung waktu eksekusi
start_time = time.time()
distances, path_matrix = bellman_ford(edges, num_nodes, start_node)
end_time = time.time()

# Menampilkan hasil
if distances:
    print("\nJarak Terpendek dari", nodes[start_node])
    for i in range(num_nodes):
        if distances[i] == INF:
            print(f"Ke {nodes[i]}: Tidak ada jalur")
        else:
            route = get_path(path_matrix, start_node, i)
            route_str = " → ".join([nodes[n] for n in route])
            print(f"Ke {nodes[i]}: {distances[i]} km, Rute: {route_str}")

    print(f"\nWaktu Eksekusi Bellman-Ford: {end_time - start_time:.6f} detik")

# Visualisasi Graph
def visualize_graph(edges, nodes):
    G = nx.Graph()

    # Tambahkan node
    for node in nodes:
        G.add_node(node)

    # Tambahkan edge
    for u, v, weight in edges:
        G.add_edge(nodes[u], nodes[v], weight=weight)

    pos = nx.spring_layout(G)
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10, font_weight='bold', arrowsize=20)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    plt.title("Visualisasi Graf Jalur Wisata")
    plt.show()

visualize_graph(edges, nodes)
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Daftar tempat wisata dan stasiun
places = [
    'Stasiun Purwokerto', 'Alun Alun Purwokerto', 'Menara Pandang Teratai',
    'Museum Bank Rakyat Indonesia', 'Museum Panglima Besar TNI Jenderal Soedirman',
    'Taman Mas Kemambang', 'Lokawisata Baturraden', 'Embung Banteran',
    'Wisata Pagubugan Melung', 'Wana Wisata Baturraden'
]

# Matriks jarak antar tempat (dalam kilometer atau satuan lain)
INF = float('inf')
distance_matrix = np.array([
    [0.0, 1.2, 3.8, 0.75, 3.9, 2.6, 13.6, 11.2, 12.3, 13.9],  # Stasiun Purwokerto
    [1.2, 0.0, 5.1, 0.65, 4.3, 2.1, 13.2, 10.8, 12.8, 13.5],  # Alun Alun Purwokerto
    [3.8, 5.1, 0.0, 5.4, 7.6, 5.9, 17.9, 15.4, 17.8, 18.2],   # Menara Pandang Teratai
    [0.75, 0.65, 5.4, 0.0, 3.7, 2.6, 13.7, 11.3, 12.7, 14.0],  # Museum Bank Rakyat Indonesia
    [3.9, 4.3, 7.6, 3.7, 0.0, 6.4, 17.3, 14.9, 12.3, 17.6],   # Museum Panglima Besar TNI Jenderal Soedirman
    [2.6, 2.1, 5.9, 2.6, 6.4, 0.0, 12.0, 8.7, 12.2, 12.4],    # Taman Mas Kemambang
    [13.6, 13.2, 17.9, 13.7, 17.3, 12.0, 0.0, 9.3, 5.0, 1.5],  # Lokawisata Baturraden
    [11.2, 10.8, 15.4, 11.3, 14.9, 8.7, 9.3, 0.0, 11.6, 8.3],  # Embung Banteran
    [12.3, 12.8, 17.8, 12.7, 12.3, 12.2, 5.0, 11.6, 0.0, 5.5],  # Wisata Pagubugan Melung
    [13.9, 13.5, 18.2, 14.0, 17.6, 12.4, 1.5, 8.3, 5.5, 0.0]    # Wana Wisata Baturraden
])

# Implementasi Algoritma Floyd-Warshall
def floyd_warshall(dist_matrix):
    n = len(dist_matrix)
    dist = np.copy(dist_matrix)
    next_node = np.full((n, n), -1, dtype=int)  # Inisialisasi jalur

    # Menyimpan jalur terpendek
    for i in range(n):
        for j in range(n):
            if dist[i][j] != INF and i != j:
                next_node[i][j] = j

    # Perhitungan jarak terpendek menggunakan Floyd-Warshall
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][j] > dist[i][k] + dist[k][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
                    next_node[i][j] = next_node[i][k]

    return dist, next_node

# Fungsi untuk melacak jalur terpendek
def get_shortest_path(next_node, start, end):
    if next_node[start][end] == -1:  # Jika tidak ada jalur
        return []

    path = [start]
    while start != end:
        start = next_node[start][end]
        path.append(start)

    return path

# Menerapkan Floyd-Warshall untuk menghitung jarak dan jalur
distances, next_node = floyd_warshall(distance_matrix)

# Menampilkan jarak terpendek antara Stasiun Purwokerto ke tempat wisata
print("=============================================")
print("Jarak Terpendek (dalam km) dari Stasiun Purwokerto:")
print("=============================================")
for i in range(len(places)):
    if distances[0][i] != INF:
        print(f"Ke {places[i]}: {distances[0][i]} km")
print("\n=============================================")

# Menampilkan jalur terpendek
print("=============================================")
print("Jalur Terpendek dari Stasiun Purwokerto ke Tempat Wisata:")
print("=============================================")
for i in range(1, len(places)):  # Mulai dari tempat selain Stasiun Purwokerto
    path = get_shortest_path(next_node, 0, i)  # 0 adalah index Stasiun Purwokerto
    path_names = [places[k] for k in path]
    print(f"Ke {places[i]}: {' -> '.join(path_names)}")

# Visualisasi graf menggunakan networkx
G = nx.Graph()
for i in range(len(places)):
    for j in range(i + 1, len(places)):
        if distance_matrix[i][j] != INF:
            G.add_edge(places[i], places[j], weight=distance_matrix[i][j])

# Menyusun posisi dengan spring layout agar graf lebih rapi
pos = nx.spring_layout(G, seed=42, k=0.3)  # seed untuk reproduktifitas dan k untuk jarak antar node

# Visualisasi graf
plt.figure(figsize=(12, 10))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", edge_color='gray', width=2)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=8)

plt.title("Graf Jalur Wisata di Purwokerto", fontsize=15)
plt.axis('off')  # Menonaktifkan axis untuk tampilan yang lebih bersih
plt.show()
```

```python
def check_if_direct_route_is_possible(start_place, end_place):
    # Mendapatkan indeks lokasi asal dan tujuan
    start_idx = list(locations.keys()).index(start_place)
    end_idx = list(locations.keys()).index(end_place)

    # Jarak langsung antar dua lokasi
    direct_distance = distance_matrix[start_idx][end_idx]

    # Jarak terpendek menggunakan Floyd-Warshall
    shortest_distance = distances[start_idx][end_idx]

    # Jika jarak langsung lebih besar daripada jarak terpendek, maka jalur langsung tidak bisa dilalui
    if direct_distance > shortest_distance:
        print(f"Untuk menuju dari {start_place} ke {end_place}, Anda harus melewati beberapa tempat terlebih dahulu.")
        return True
    else:
        print(f"Jalur langsung dari {start_place} ke {end_place} dapat dilalui.")
        return False

# Cek beberapa pasangan lokasi untuk mengetahui apakah jalur langsung bisa dilalui
for start in locations.keys():
    for end in locations.keys():
        if start != end:
            check_if_direct_route_is_possible(start, end)
```
